```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from math import inf
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(pd.read_csv(''))
Data_Set_np = df.to_numpy()
Data_Set_np = Data_Set_np.astype(float)
X = Data_Set_np[:, :-6]
y = Data_Set_np[:, -6:]
```

```python
def run_pca(X):
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    for i in range(X.shape[1]):
        pca = PCA(n_components=i) # Example: reduce to 2 components
        pca.fit(X_scaled)
        X_pca = pca.transform(X_scaled)
        print("features = ", i)
        print(pca.explained_variance_ratio_)
        print(pca.explained_variance_ratio_.sum()) # Cumulative explained variance
    return X_pca
```

```python
def pass_tests(y_inst):
    for test in y_inst:
        if test is None:
            return False
    return True

def find_valid(X,y):
    X_new = []
    y_new = []
    for i in range(X.shape[0]):
        if pass_tests(y[i]):
            X_new.append(X[i])
            y_new.append(y[i])
    return X_new, y_new
```
